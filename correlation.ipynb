```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# drop records with null values

df1 = df.dropna()
df1.isnull().sum()
```

```python
# Convert data type of columns to appropriate types

df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
df1
```

```python
# filt = (df1['country'] == 'USA') & (df1['writer'] == 'Brian Azzarello')
# filt
```

```python
df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# create correct year column
# df1['year'] = df1['released'].astype(str).str[:4]

# sort values by gross
df1 = df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates
df1.drop_duplicates()
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df1['budget'], y=df1['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Movies')

plt.show
```

```python
correlation_matrix = df1.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

### Company may have a high correlation. Lets find out if it is true or not.

```python
df1.head()
```

```python
df_encoded = df1

for col_name in df_encoded.columns:
    if(df_encoded[col_name].dtype == 'object'):
        df_encoded[col_name] = df_encoded[col_name].astype('category')
        df_encoded[col_name] = df_encoded[col_name].cat.codes

df_encoded
```

```python
correlation_matrix = df_encoded.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_encoded.corr()
```

```python
# unstack

correlation_mat = df_encoded.corr()
corr_pairs  = correlation_mat.unstack()
corr_pairs
```

```python
# sort pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
